###  Gráficos de contorno

Para representar funções em 2D, $v(x,y)$ podemos usar gráficos de contorno, que desenham linhas em que o valor de $v(x,y)$ é constante. No matplotlib temos duas funções  **contour(x,y,v)** e  **contourf(x,y,v)** para o fazer.
Uma ilustração simples está na célula seguinte. Podes trocar entre estes dois casos alterando as linhas comentadas.

```python
fig=plt.figure(figsize=(8,8))                    # cria uma figura de tamanho 8x8
ax = fig.add_subplot(111,aspect='equal')         # mesma escala para x e y

def V(x,y):                                      # potencial de dipolo
    r2=x**2+y**2
    return y/r2
    
max_x =.1
max_y = .1

linhas=np.linspace(-40,40,21)         # determina valores da função para os contornos a 
                                      # contornos a representar

                                      # definição da grelha de pontos onde a função é
points=40                             # calculada
x = np.linspace(-max_x,max_x,points) 
y = np.linspace(-max_y,max_y, points)
X,Y = np.meshgrid(x,y)
Z1 = V(X,Y)                             # lista de valores da função

#contour linhas
CS = ax.contour(X, Y, Z1,levels=linhas)
plt.title('Dipole  potencial', fontsize=18);
plt.colorbar(CS);
# countor filled
#CS = ax.contourf(X, Y, Z1,levels=linhas)
plt.grid(True)
```
