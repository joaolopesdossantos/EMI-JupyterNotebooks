# Demos das teóricas


### Comparação campo de fio e de linha de cargas

```python
# campo E_z de N cargas em linha

#  campo de fio
def Ez(z):
    if z>1/2:
        return 1/(z**2-1/4)
    if z<-1/2:
        return -1/(z**2-1/4)
    return z/(1/4-z**2 )


# campo de linha de cargas
def Ez_discreto (z, N):
    charge=1/N
    d=1/N
    field=0
    for i in range(N):
        field+=charge/(z+ 1/2-i*d)**2
    return field 
#
N=10                                     # número de cargas

#                                       
zpts=100                                                                 
z=np.linspace(.501,2,zpts)              # coordenadas para gráfico
ezc=np.zeros(zpts, float)               # arrays para valores de campos
ezd1=np.zeros(zpts,float)
ezd2=np.zeros(zpts,float)

for i in range(zpts):                   # valores de campo de fio
    ezc[i]=Ez(z[i])
#
for i in range(zpts):                   # valores de campo discreto
    ezd1[i]=Ez_discreto(z[i], N)
#
for i in range(zpts):
    ezd2[i]=Ez_discreto(z[i], 10*N)
                                         # Gráficos
fig=plt.figure(figsize=(12,8))   
line1,line2,line3,=plt.plot(z,ezd1,z,ezd2,z,ezc)
plt.grid(True)
plt.xlabel(r'$z$', fontsize=16 )
plt.ylabel(r'$E_z$', fontsize=16)
plt.legend((line1,line2,line3), (r'$N=10$',r'$N=100$',r'$N=\infty$'))
plt.axis([.501,2,0,20])
```

### Elemento de área em coordenadas polares

```python
fig=plt.figure(figsize=(12,12))                    # cria uma figura de tamanho 8x8
ax = fig.add_subplot(111,aspect='equal')
r0=1

q0= np.pi/4
                            #change these parameters to make the area element smaller
dr=1                        # delta r
dq=1                        # delta theta 
#
r=np.linspace(r0,r0+dr,100)
qs=np.linspace(q0,q0+dq,100)
plt.plot(r*np.cos(q0),r*np.sin(q0),'b-')
plt.plot(r*np.cos(q0+dq),r*np.sin(q0+dq),'r-')
plt.plot(r0*np.cos(qs),r0*np.sin(qs))
plt.plot((r0+dr)*np.cos(qs),(r0+dr)*np.sin(qs))
plt.grid(True)
#plt.plot(0,0,'ko')
#plt.axis([-.5,1.5,-.1,2])
```
