Uma maneira padrão de comparar as intensidades da interação eletromagnética e gravítica é calcular a razão das forças gravíticas e eletricas entre duas partículas elementares, por exempo, dois electrões. Começa por definir os valores SI das constantes fundamentais como por exemplo, 

```python
G = 6.67430E-11
```

Depois determina em termos das constantes universais  a razão indicada e comenta o resultado. 

```python
# calculo da razão FG/FE
G = 6.67430E-11
k=8.988E9
me=9.109E-31
e=1.602E-19
FG=G*me*me
FC=k*e*e
FG/FC
```

### Gráfico de Campo
Pretende-se representar o valor d campo elétrico criado por duas cargas $q_1=2\mu  C$ e $q_2=-2\mu C$ distantes de  $d=1$ m no eixo $Ox$ ao longo da linha que bissecta o segmento que une as duas cargas. Escreve uma função que calcule o valor desse campo e representa-o graficamente. 
O valor do campo é 
$$
E_y(z)=2\frac{q}{4\pi\epsilon_0}\frac{d/2}{(d^2/4+z^2)^{3/2}}
$$

```python
def Ex(y):
    k=9E9
    q=2E-6
    d=1
    E1=2*k*q*(d/2)/(d**2/4+y**2)**1.5
    return E1
y=np.linspace(-5,5,100)
field=Ex(y)/1E3                               # campo em kV/m
plot(y,field)
grid(True)
ylabel(r'$E_x/ 10^3$ V/m', fontsize=16)
xlabel (r'$y$ /m', fontsize=16);
```
